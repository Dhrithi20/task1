```python
import pandas as pd
df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df.shape         
df.head()        
df.info()        
df.isnull().sum()  
```

```python
df = df.drop_duplicates()
print("After removing duplicates:", df.shape)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns 
```

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')

# Drop rows where important info is missing
df = df.dropna(subset=['country', 'date_added', 'rating'])

print("After handling missing values:", df.shape)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
text_cols = ['type', 'title', 'director', 'cast', 'country', 'rating', 'listed_in', 'description']

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
df.info()     
df.head()     
```

```python
df.to_csv("netflix_titles_cleaned.csv", index=False)
```
